```python
%load_ext autoreload
%autoreload 2
import os
import numpy as np
import pandas as pd
from src.classes.ClassFractionalLorenz96 import FractionalLorenz96
import matplotlib.pyplot as plt
from configs.RecursiveVPSDE.Markovian_8DLorenz.recursive_Markovian_PostMeanScore_8DLorenz_Stable_T256_H05_tl_110data_StbleTgt import get_config
config = get_config()
```

```python
num_paths = 100
t0 = config.t0
t1 = int(10*config.t1)
deltaT = config.deltaT
initial_state = np.array(config.initState)
rvs = None
H = 0.5
ts_length = int(5*config.ts_length)
fLnz = FractionalLorenz96(X0=config.initState, diff=config.diffusion, num_dims=config.ndims,
                                      forcing_const=config.forcing_const)
data = np.array(
    [fLnz.euler_simulation(H=H, N=ts_length, deltaT=deltaT, X0=initial_state, Ms=None, gaussRvs=rvs,
                           t0=t0, t1=t1) for _ in (range(num_paths))]).reshape(
    (2, num_paths, ts_length + 1, config.ndims))
drift = data[1,:,:,:]
data = data[0,:,:,:]
cummean = np.cumsum(np.nanmean(np.sum(np.power(data,2), axis=-1), axis=0)) / np.arange(1, data.shape[1] + 1)
```

```python
fig, ax = plt.subplots(figsize=(14,8))
ts = np.linspace(1/256, t1, data.shape[1])
plt.yscale("log")
ax.tick_params(labelsize=36)
plt.plot(ts, cummean, linewidth=4, label="True CumMean")
plt.xlabel("Time axis ($t_j$)", fontsize=36)
plt.ylabel("Value", fontsize=36)
plt.legend(fontsize=35)
plt.show()
plt.close()
```
