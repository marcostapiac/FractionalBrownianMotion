```python
import numpy as np
import os
import matplotlib.pyplot as plt
from configs import project_config
```

```python
num_paths = 1024
save_path = (
            project_config.ROOT_DIR + f"experiments/results/8DLnz_NewDriftEvalExp_MSEs_{num_paths}NPaths").replace(
    ".", "")
all_true_paths=np.load(save_path+"_true_paths.npy", allow_pickle=True)
all_true_paths = all_true_paths.reshape((-1, all_true_paths.shape[-2], all_true_paths.shape[-1]), order="C")
all_score_paths=np.load(save_path+"_score_paths.npy", allow_pickle=True)
all_score_paths = all_score_paths.reshape((-1, all_score_paths.shape[-2], all_score_paths.shape[-1]), order="C")
all_nad_paths=np.load(save_path+"_nad_paths.npy", allow_pickle=True)
all_nad_paths = all_nad_paths.reshape((-1, all_nad_paths.shape[-2], all_nad_paths.shape[-1]), order="C")
is_prevPath_obs=np.load(save_path+"_trainingPrevPath_paths.npy", allow_pickle=True)
is_prevPath_incs=np.load(save_path+"_trainingIncs_paths.npy", allow_pickle=True)
true_drift=np.load(save_path+"_true_drifts.npy", allow_pickle=True)
true_drift = true_drift.reshape((1000, 256, 8), order="C")
all_score_drift_ests_true_law=np.load(save_path+"_score_drifts.npy", allow_pickle=True)
all_score_drift_ests_true_law = all_score_drift_ests_true_law.reshape((1000, 256, 8), order="C")
all_nad_drift_ests_true_law=np.load(save_path+"_nad_drifts.npy", allow_pickle=True)
all_nad_drift_ests_true_law = all_nad_drift_ests_true_law.reshape((1000, 256, 8), order="C")
```

```python
tidx = 256//2
for d in range(8):
    fig, ax = plt.subplots(figsize=(14, 8))
    plt.hist(true_drift[:, tidx, d], density=False, label="True Drift")
    plt.hist(all_nad_drift_ests_true_law[:, tidx, d], density=False, label="Nad Drift")
    plt.title(f"Drift at time {tidx} for dimension {d+1}", fontsize=40)
    plt.legend()
    plt.show()
    plt.close()
```
